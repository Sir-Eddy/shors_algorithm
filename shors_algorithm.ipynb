# Introduction and Motivation

The integrity and confidentiality of modern digital communication infrastructure rely fundamentally on the security of asymmetric cryptosystems. In particular, the RSA scheme, whose security depends on the mathematical complexity of prime factorization of large integers, forms the backbone of numerous protocols, ranging from securing web traffic (TLS/SSL) to digital signatures and financial transactions. In classical computer science, this mathematical problem is considered practically intractable, as the resource requirements for the best known classical algorithms, such as the General Number Field Sieve (GNFS), grow sub-exponentially with the key length. However, this security model, considered robust for decades, has been challenged by theoretical breakthroughs in quantum theory.

In 1994, Peter Shor formulated a quantum algorithm capable of solving prime factorization and the related discrete logarithm problem in polynomial time (Shor, 1997). By exploiting quantum mechanical phenomena, specifically superposition and entanglement, Shor's algorithm offers an exponential speedup over classical methods. Theoretically, this implies that a sufficiently powerful quantum computer could break current RSA key lengths, such as 2048 or 4096 bits, within a timeframe relevant for cryptographic attacks.

Although the theoretical correctness of the algorithm is mathematically proven, its practical application faces immense physical and engineering barriers. The discrepancy between the theoretical algorithmic formulation and physical reality is severe. While the algorithm assumes ideal qubits and error-free gate operations, real quantum systems are easily influenced by decoherence and noise caused by environmental interactions. Consequently, implementing Shor's algorithm for cryptographically relevant numbers requires not only an enormous number of physical qubits but also precise control over these systems for durations far exceeding current coherence times. Furthermore, the necessity for robust quantum error correction significantly increases resource demands and drastically exacerbates hardware requirements.

Against this background, the present thesis is dedicated to a detailed examination of the challenges and advances in the physical implementation of Shor's algorithm. The objective is to analyze the gap between algorithmic theory and the current hardware reality. In doing so, both fundamental physical requirements—such as gate fidelity, qubit connectivity, and coherence times—and the specific hurdles of various technological platforms, from superconducting circuits to ion traps, will be discussed. By evaluating historical experiments and current developments, this work aims to assess how far research has progressed toward a scalable realization and which technological milestones remain to be achieved before the theoretical threat posed by Shor's algorithm becomes a physical reality.
